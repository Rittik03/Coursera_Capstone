```python
import pandas as pd 
import numpy as np 
```

```python
India_population = pd.read_csv(r'C:\Users\ritti\Desktop\in.csv')
```

```python
India_population.head()
```

```python
India_cities = India_population.drop(['capital', 'iso2', 'country', 'admin_name', 'population' , 'population_proper'] , axis= 1)
```

```python
India_cities.head()
```

```python
import folium 
from geopy.geocoders import Nominatim
```

```python
address =  'India'
geolocator = Nominatim(user_agent="ind_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
map_india = folium.Map(location=[latitude, longitude], zoom_start=4)
map_india
```

```python
for lat, lng, city,  in zip(India_cities['lat'], India_cities['lng'], India_cities['city']):
    label = city
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_india)  
    
map_india
```

```python
main_hub_lat = India_cities['lat'].mean()
mani_hub_lang = India_cities['lng'].mean()
```

```python
print('The latitude of the main-hub airport will be {} , and the longitude of the main-hub airport will be {}'.format(main_hub_lat, mani_hub_lang))
```
